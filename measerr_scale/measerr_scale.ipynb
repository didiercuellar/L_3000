```python
import pandas as pd 
import numpy as np 
import scipy.stats as st 
from scipy.interpolate import interp1d
from matplotlib import pyplot as plt 
import math
import statistics as stats
```

```python
data=pd.read_csv('measerr_scale_U.csv',header=0,)
data = data.dropna(how='any')#quita los valores nan del documento
y=data.iloc[:,0]
w=data.iloc[:,1]
x=data.iloc[:,2]

print(stats.stdev(x))
print(data)
```

```python
yerror = w
xerror = 1.4858944399056223e+45
plt.plot(x,y,'ro')
plt.errorbar(x,y,yerr=yerror, fmt = ' ' , ecolor='black')
plt.errorbar(x,y,xerr=xerror, fmt = ' ' , ecolor='blue')
plt.ylabel('measerr_scale_U')
plt.xlabel('erg/s')
plt.title("L_3000_U")
plt.xscale('log')
#plt.yscale('log')
plt.grid()
plt.savefig("L_3000_measerr_scale_U_error.pdf")
plt.show
```
